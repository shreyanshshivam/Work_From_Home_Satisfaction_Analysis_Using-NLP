## Classification 

```python
import pandas as pd
import numpy as np
import json
import collections 
import matplotlib.pyplot as plt
import itertools

import re
import nltk
import gensim
```

### prepare training data

```python
# get training data

col_list= ['topic','advert_label']
train_all_a = pd.read_csv("sample_train_class(final).csv")
train_a = pd.read_csv("sample_train_class(final).csv",usecols=col_list)


train_all_b = pd.read_csv("sample_train_class2(final).csv")
train_b = pd.read_csv("sample_train_class2(final).csv",usecols=col_list)

comb = [train_a, train_b]
comb_all = [train_all_a, train_all_b]

comb_train = pd.concat(comb)
comb_train_all=pd.concat(comb_all)

print(len(comb_train))

#comb_train_all.to_csv('sample_traincombined_class(final).csv',index=False)
```

```python
# to see full column content instead of truncated text
pd.set_option('display.max_colwidth', -1)

# first x rows
print(comb_train.head(10))
```

```python
train_text  = comb_train['topic']
print(train_text.head())
```

```python
#code to convert df to list 

token = comb_train['topic'].tolist()
#print(token)

def convert(lst): 
    return eval(lst)

final=[]
for i in token:
    tokenf=convert(i)
    final.append(tokenf)

print(final)
```

```python
#Code to covert docs to vec
def docs2vecs(docs, dictionary):
    vecs1 = [dictionary.doc2bow(doc) for doc in docs]
    return vecs1
```

```python
#create dictionary and vectors for documents

from gensim import corpora
tweet_dictionary = gensim.corpora.Dictionary(final)
tweet_vecs = docs2vecs(final, tweet_dictionary)

print(tweet_dictionary)
print(tweet_vecs[2])
```

```python
## save dictionary 

tweet_dictionary.save('class_dict')
```

```python
#covert label into list

label=comb_train['advert_label'].tolist()
#print(label)
```

```python
#Label the trained data. 

#convert to dict with the term ids given frequency 1
all_data_as_dict = [{id:1 for (id, tf_value) in vec} for vec in tweet_vecs]
print(type(all_data_as_dict))
print (all_data_as_dict[0])

#create tuples
labeled_data = list(zip(all_data_as_dict, label))

print(labeled_data[5])
```

### build NLTK NaiveBayes classifier as a trial

```python
#Generate the trained classifier
classifier = nltk.NaiveBayesClassifier.train(labeled_data)
```

### prepare test data

```python
# get test data

col_list= ['topic','advert_label']
test_all = pd.read_csv("sample_test_class(final).csv")

test= pd.read_csv("sample_test_class(final).csv",usecols=col_list)

print(len(test_all))
print(len(test_all))
```

```python
test_text  = test['topic']
print(test_text.head())
```

```python
#code to convert df to list 

token1 = test['topic'].tolist()
#print(token)

def convert(lst): 
    return eval(lst)

final1=[]
for i in token1:
    tokenf=convert(i)
    final1.append(tokenf)

print(final1)
```

```python
#create  vectors for documents uusing dictionary in training dataset

tweet_vecs_test = docs2vecs(final1, tweet_dictionary)

print(tweet_vecs_test[0])
```

```python
#covert label into list

label_test=test['advert_label'].tolist()
#print(label_test)
```

```python
#Label the test data. 

#convert to dict with the term ids given frequency 1
all_data_as_dict1 = [{id:1 for (id, tf_value) in vec} for vec in tweet_vecs_test]
print(type(all_data_as_dict1))
print (all_data_as_dict1[0])

#create tuples
labeled_data_test = list(zip(all_data_as_dict1, label_test))

print(labeled_data_test[0])
```

```python
#check performance
print(nltk.classify.accuracy(classifier, labeled_data_test))
```

```python
## get precision, recall etc. 

import sklearn

test_result = []
gold_result = []

for i in range(len(labeled_data_test)):
    test_result.append(classifier.classify(labeled_data_test[i][0]))
    gold_result.append(labeled_data_test[i][1])

from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix   

print('\nClasification report:\n', classification_report(gold_result, test_result))
print('\nConfussion matrix:\n',confusion_matrix(gold_result, test_result))
```

### Try different classifier methods

```python
from nltk.classify.scikitlearn import SklearnClassifier

from sklearn.naive_bayes import MultinomialNB,BernoulliNB

from sklearn.linear_model import LogisticRegression,SGDClassifier
from sklearn.svm import SVC, LinearSVC



print("Original Naive Bayes Algo accuracy percent:", (nltk.classify.accuracy(classifier, labeled_data_test))*100)
#classifier.show_most_informative_features(15)

MNB_classifier = SklearnClassifier(MultinomialNB())
MNB_classifier.train(labeled_data)
print("MNB_classifier accuracy percent:", (nltk.classify.accuracy(MNB_classifier, labeled_data_test))*100)

BernoulliNB_classifier = SklearnClassifier(BernoulliNB())
BernoulliNB_classifier.train(labeled_data)
print("BernoulliNB_classifier accuracy percent:", (nltk.classify.accuracy(BernoulliNB_classifier, labeled_data_test))*100)

LogisticRegression_classifier = SklearnClassifier(LogisticRegression())
LogisticRegression_classifier.train(labeled_data)
print("LogisticRegression_classifier accuracy percent:", (nltk.classify.accuracy(LogisticRegression_classifier, labeled_data_test))*100)

SGDClassifier_classifier = SklearnClassifier(SGDClassifier())
SGDClassifier_classifier.train(labeled_data)
print("SGDClassifier_classifier accuracy percent:", (nltk.classify.accuracy(SGDClassifier_classifier, labeled_data_test))*100)

SVC_classifier = SklearnClassifier(SVC())
SVC_classifier.train(labeled_data)
print("SVC_classifier accuracy percent:", (nltk.classify.accuracy(SVC_classifier, labeled_data_test))*100)

LinearSVC_classifier = SklearnClassifier(LinearSVC())
LinearSVC_classifier.train(labeled_data)
print("LinearSVC_classifier accuracy percent:", (nltk.classify.accuracy(LinearSVC_classifier, labeled_data_test))*100)

```

```python
final_classifier = BernoulliNB_classifier.train(labeled_data)
```
