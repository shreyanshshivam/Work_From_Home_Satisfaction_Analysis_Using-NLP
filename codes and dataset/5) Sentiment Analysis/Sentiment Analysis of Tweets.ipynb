```python
import pandas as pd
import sqlite3
import json
import matplotlib.pyplot as plt
import re
import seaborn as sns
import nltk
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from collections import Counter
```

```python
data = pd.read_csv("combined_data_classified(non_advert).csv")
print(len(data))
print(data['sentiment'].head())
```

### run text blob

```python
desc_blob = [TextBlob(desc) for desc in data['sentiment']]
```

```python
data['tb_Polarity'] = [b.sentiment.polarity for b in desc_blob]
data['tb_Subjectivity'] = [b.sentiment.subjectivity for b in desc_blob]
```

### run vaderSentiment

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
data['compound'] = [analyzer.polarity_scores(v)['compound'] for v in data['sentiment']]
data['neg'] = [analyzer.polarity_scores(v)['neg'] for v in data['sentiment']]
data['neu'] = [analyzer.polarity_scores(v)['neu'] for v in data['sentiment']]
data['pos'] = [analyzer.polarity_scores(v)['pos'] for v in data['sentiment']]
```

```python
def ratio(x):
 if x > 0:
    return 'positive'
 elif x == 0:
        return 'neutral'
 else:
    return 'negative'
```

```python
data['tb_analysis'] = data['tb_Polarity'].apply(ratio)
data['vd_analysis'] = data['compound'].apply(ratio)
```

```python
pd.DataFrame(data).to_csv("NonAdvert_Sent_Output.csv",index=False)
```
